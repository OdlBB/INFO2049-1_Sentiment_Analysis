```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# GRU vs LSTM

```python
lstm_acc = 0.8654
lstm_recall = 0.8836
lstm_prec = 0.8439

gru_acc = 0.8598
gru_recall = 0.88
gru_prec = 0.8356
```

```python
print("-----------------------------------------------------------------------------------")
print("                      Accuracy               Precision                   Recall    ")
print("-----------------------------------------------------------------------------------")
print(" LSTM              ", lstm_acc, "                   ", lstm_prec, "                 ", lstm_recall)
print(" GRU               ", gru_acc, "                   ", gru_prec, "                  ", gru_recall)
print("-----------------------------------------------------------------------------------")
```

```python
results = ['Accuracy', 'Precision', 'Recall']
lstm = [lstm_acc, lstm_prec, lstm_recall]
gru = [gru_acc, gru_prec, gru_recall]

x = np.arange(len(results))
width = 0.3

fig, ax = plt.subplots(figsize=(10, 7))
bar1 = ax.bar(x - width/2, lstm, width, label='LSTM', color='dodgerblue')
bar2 = ax.bar(x + width/2, gru, width, label='GRU', color='darkred')

for rect in bar1 + bar2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')

ax.set_xticks(x)
ax.set_xticklabels(results)
ax.legend()



plt.show()

```

# Word embeddings

```python
word2vec_pretrained_acc = 0.854
word2vec_pretrained_rec = 0.8884
word2vec_pretrained_prec = 0.8122

word2vec_retrained_acc = 0.8674
word2vec_retrained_rec = 0.8296
word2vec_retrained_prec = 0.9273

glove_acc = 0.8588
glove_rec = 0.8807
glove_prec = 0.8324

fasttext_acc = 0.8616
fasttext_rec = 0.8782
fasttext_prec = 0.842
```

```python
print("-----------------------------------------------------------------------------------------------------")
print("                                        Accuracy               Precision                   Recall    ")
print("-----------------------------------------------------------------------------------------------------")
print(" Pretrained word2vec              ", word2vec_pretrained_acc, "                     ", word2vec_pretrained_prec, "                 ", word2vec_pretrained_rec)
print(" Retrained word2vec               ", word2vec_retrained_acc, "                    ", word2vec_retrained_prec, "                 ", word2vec_retrained_rec)
print(" Glove                            ", glove_acc, "                    ", glove_prec, "                 ", glove_rec)
print(" FastText                         ", fasttext_acc, "                    ", fasttext_prec, "                  ", fasttext_rec)
print("-----------------------------------------------------------------------------------------------------")
```

```python
pretrained_word2vec = [word2vec_pretrained_acc, word2vec_pretrained_prec, word2vec_pretrained_rec]
retrained_word2vec = [word2vec_retrained_acc, word2vec_retrained_prec, word2vec_retrained_rec]
glove = [glove_acc, glove_prec, glove_rec]
fasttext = [fasttext_acc, fasttext_prec, fasttext_rec]

x = np.arange(len(results))
width = 0.15

fig, ax = plt.subplots(figsize=(20, 7))
bar1 = ax.bar(x - 3*width/2, pretrained_word2vec, width, label='Pre-trained Word2Vec', color='dodgerblue')
bar2 = ax.bar(x - width/2, retrained_word2vec, width, label='Re-trained Word2Vec', color='darkred')
bar3 = ax.bar(x + 3*width/2, glove, width, label='Glove', color='darkgreen')
bar4 = ax.bar(x + width/2, fasttext, width, label='FastText', color='darkblue')

for rect in bar1 + bar2 + bar3 + bar4:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')

ax.set_xticks(x)
ax.set_xticklabels(results)
ax.legend()



plt.show()
```
